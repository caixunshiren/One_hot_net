```python
import matplotlib.pyplot as plt
import numpy as np
```

### Accuracy VS Dimension Multiplier for Encoder Layer Only MLP

```python
multipliers = np.array([12,20,40,80,160])
w_threshold_acc = np.array([96.8, 97.3, 97.5, 97.6, 97.8])
w_o_threshold_acc = np.array([89,88,88,87,88])
```

```python
plt.plot(multipliers, w_threshold_acc, 'o-', color='lightblue',label = "With Threshold")
plt.plot(multipliers, w_o_threshold_acc, 'o-', color='red', label = "Without Threshold")
plt.legend()
plt.title("Accuracy VS Dimension Multiplier for Encoder Layer Only MLP")
plt.xlabel("Dimension Multiplier")
plt.ylabel("% Accuracy")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Accuracy VS Dimension Multiplier for Encoder Layer Only MLP

```python
multipliers = np.array([12,20,40,80,160])
w_threshold_acc = np.array([96.8, 97.3, 97.5, 97.6, 97.8])
w_o_threshold_acc = np.array([89,88,84,89,90])
```

```python
plt.plot(multipliers, w_threshold_acc, linestyle = 'dotted', label = "With Threshold")
plt.plot(multipliers, w_o_threshold_acc, linestyle = 'dotted', label = "Without Threshold")
plt.legend()
plt.title("Accuracy VS Dimension Multiplier for Encoder Layer Only MLP")
plt.show()
```
